## Implement your own median filtering algorithm using a mask of size 3*3.

```python
def median_filter(image):
    height = image.shape[0]
    width = image.shape[1]
    filtered_image = np.zeros([height,width])
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            neighborhood = image[i:i + 3, j:j + 3]
            median_value = np.median(neighborhood)
            filtered_image[i, j] = median_value
            
    return filtered_image
```

```python
def skimage_median_filter(image):
    skimage_median_filter_image =  filters.median(image, footprint=[[1, 1, 1],
 [1, 1, 1],
 [1, 1, 1]])
    return skimage_median_filter_image
```

```python
def skimage_gaussion_filter(image, sigma):
    skimage_gaussion_filter_image =  filters.gaussian(image, sigma)
    return skimage_gaussion_filter_image
```
